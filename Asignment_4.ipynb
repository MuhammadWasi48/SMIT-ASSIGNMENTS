1. Create a class 'BankAccount' with attributes account holder name, account number, and
balance. Add methods for deposit and withdrawal. Implement logic to prevent withdrawal if balance
is insufficient.

```python
class BankAccount:
    def __init__(self, holder_name, account_number, balance):
        self.holder_name = holder_name
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited Amount:", amount)
        print("Current Balance:", self.balance)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn Amount:", amount)
            print("Remaining Balance:", self.balance)
        else:
            print("Insufficient Balance! Withdrawal not allowed.")

# Creating an object
account = BankAccount("Safi", "123456789", 8000)

# Deposit money
account.deposit(45000)

# Withdraw money
account.withdraw(3500)

# Try to withdraw more than balance
account.withdraw(1600)
```

2. Design a class 'Student' with attributes name and marks. Create a method to calculate grade
based on marks using conditions (A, B, C, Fail).

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def calculate_grade(self):
        if self.marks >= 80:
            grade = "A"
        elif self.marks >= 60:
            grade = "B"
        elif self.marks >= 40:
            grade = "C"
        else:
            grade = "Fail"

        print("Student Name:", self.name)
        print("Marks:", self.marks)
        print("Grade:", grade)

# Creating an object
student1 = Student("Safi", 92)

# Calling method
student1.calculate_grade()
```

3. Create a class 'ElectricityBill' for a KE-electric consumer. Include units consumed and calculate
bill based on slab rates (e.g., different price per unit for different ranges).

```python
class ElectricityBill:
    def __init__(self, consumer_name, units):
        self.consumer_name = consumer_name
        self.units = units

    def calculate_bill(self):
        if self.units <= 100:
            bill = self.units * 10
        elif self.units <= 200:
            bill = (100 * 10) + ((self.units - 100) * 15)
        elif self.units <= 300:
            bill = (100 * 10) + (100 * 15) + ((self.units - 200) * 20)
        else:
            bill = (100 * 10) + (100 * 15) + (100 * 20) + ((self.units - 300) * 25)

        print("Consumer Name:", self.consumer_name)
        print("Units Consumed:", self.units)
        print("Electricity Bill: Rs.", bill)

# Creating an object
customer = ElectricityBill("Safi", 48000)

# Calculate bill
customer.calculate_bill()
```

4. Design a class 'Vehicle' and create subclasses 'Car' and 'Bike'. Override a method
'start_engine()' differently in each subclass to demonstrate polymorphism.

```python
# Parent Class
class Vehicle:
    def start_engine(self):
        print("Vehicle engine is starting...")

# Child Class 1
class Car(Vehicle):
    def start_engine(self):
        print("Car engine starts with a key or push button.")

# Child Class 2
class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine starts with a self-start or kick.")

# Creating objects
car = Car()
bike = Bike()

# Calling methods
car.start_engine()
bike.start_engine()
```

5. Create a class 'Employee' with basic salary. Add a method to calculate total salary including
bonus (10%) and tax deduction (5%).

```python
class Employee:
    def __init__(self, name, basic_salary):
        self.name = name
        self.basic_salary = basic_salary

    def calculate_salary(self):
        bonus = self.basic_salary * 0.10      # 10% Bonus
        tax = self.basic_salary * 0.05        # 5% Tax Deduction

        total_salary = self.basic_salary + bonus - tax

        print("Employee Name:", self.name)
        print("Basic Salary:", self.basic_salary)
        print("Bonus (10%):", bonus)
        print("Tax Deduction (5%):", tax)
        print("Total Salary:", total_salary)

# Creating an object
emp = Employee("Safi", 148000)

# Calling method
emp.calculate_salary()
```

6. Design an abstract class 'Shape' with a method 'area()'. Create subclasses 'Circle' and
'Rectangle' and implement area calculation logic.

```python
from abc import ABC, abstractmethod
import math

# Abstract Class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# Circle Class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

# Rectangle Class
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Creating Objects
circle = Circle(8)
rectangle = Rectangle(14 , 8)

# Display Areas
print("Area of Circle =", circle.area())
print("Area of Rectangle =", rectangle.area())
```

7. Create a class 'HospitalPatient' with attributes name, age, and disease. Add a method to
determine priority (Emergency if age > 60 or serious disease).

```python
class HospitalPatient:
    def __init__(self, name, age, disease):
        self.name = name
        self.age = age
        self.disease = disease

    def determine_priority(self):
        serious_diseases = ["Heart Attack", "Cancer", "Stroke", "Kidney Failure"]

        if self.age > 60 or self.disease in serious_diseases:
            priority = "Emergency"
        else:
            priority = "Normal"

        print("Patient Name:", self.name)
        print("Age:", self.age)
        print("Disease:", self.disease)
        print("Priority:", priority)

# Creating an object
patient = HospitalPatient("Safi", 22, "Brain Tumar")

# Calling method
patient.determine_priority()
```

8. Build a class 'ShoppingCart' where user can add items with price. Implement logic to calculate
total bill and apply discount if total exceeds a certain amount.

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item_name, price):
        self.items.append((item_name, price))

    def calculate_bill(self):
        total = 0

        for item in self.items:
            total += item[1]

        # Apply 10% discount if bill exceeds 5000
        if total > 5000:
            discount = total * 0.10
        else:
            discount = 0

        final_bill = total - discount

        print("Total Bill:", total)
        print("Discount:", discount)
        print("Final Bill:", final_bill)

# Creating object
cart = ShoppingCart()

# Adding items
cart.add_item("Shoes", 23000)
cart.add_item("Shirt", 5900)
cart.add_item("Watch", 28000)

# Calculating bill
cart.calculate_bill()
```

9. Create a class 'MobilePhone' with private attribute battery percentage. Use getter and setter
methods to update battery with validation (0–100 only).

```python
class MobilePhone:
    def __init__(self, battery_percentage):
        self.__battery_percentage = battery_percentage  # Private Attribute

    # Getter Method
    def get_battery(self):
        return self.__battery_percentage

    # Setter Method
    def set_battery(self, battery):
        if 0 < battery <= 100:
            self.__battery_percentage = battery
            print("Battery updated successfully.")
        else:
            print("Error! Battery percentage must be between 0 and 100.")

# Creating object
phone = MobilePhone(80)

# Getting battery percentage
print("Current Battery:", phone.get_battery(), "%")

# Updating battery percentage
phone.set_battery(65)
print("Updated Battery:", phone.get_battery(), "%")

# Invalid value
phone.set_battery(110)
```

10. Design a class 'Library' where users can borrow and return books. Add logic to check
availability before issuing a book.

```python
class Library:
    def __init__(self):
        self.books = ["Python", "Java", "C++", "Database"]

    # Borrow Book
    def borrow_book(self, book_name):
        if book_name in self.books:
            self.books.remove(book_name)
            print(f"{book_name} has been issued successfully.")
        else:
            print(f"Sorry! {book_name} is not available.")

    # Return Book
    def return_book(self, book_name):
        self.books.append(book_name)
        print(f"{book_name} has been returned successfully.")

    # Display Available Books
    def show_books(self):
        print("Available Books:")
        for book in self.books:
            print("-", book)


# Creating Object
library = Library()

# Display books
library.show_books()

# Borrow a book
library.borrow_book("Python")

# Check books after borrowing
library.show_books()

# Return a book
library.return_book("Python")

# Check books after returning
library.show_books()
```
