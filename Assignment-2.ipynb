# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
marks = int(input("Enter marks"))
if marks > 90 & marks < 100:
    print("A")
elif marks > 80 & marks < 100:
    print("B")
elif marks > 70 & marks < 100:
    print("C")
elif marks > 60 & marks < 100:
    print("D")
elif marks < 60 & marks > 0:
    print("Fail")    
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
User = int(input('Enter month no. (1 - 12): '))
if user == 12 or user == 2 or user == 1:
    print("Winter")
elif user == 3 or user == 4  or user == 5:
    print("Spring")
elif user == 6 or user == 7 or user == 8:
    print("Summer")
else:
    print("Autumn")        
```

# User se 2 numbers lo:

# pehla > dusra → "First is greater"

# warna → "Both equal"

```python
A = int(input("Enter no.: "))
B = int(input("Enter no.: "))
if A > B:
    print("First is greater")
if A < B:
    print("Second is greater")
if A == B:
    print("Both Equal")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
a = 4
b = 5
op = input("Enter Operator (+,-,*,/)")
if op == "+" :
    print(a+b)
elif op == "-" :
    print(a-b)
elif op == "*" :
    print(a*b)
elif op == "/" :
    print(a/b)
else:
    print("invalid operator")
```
